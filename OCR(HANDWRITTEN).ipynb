```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split
import tensorflow as tf
from tensorflow import keras
%matplotlib inline
```

```python
data=pd.read_csv('optdigits.tra',header=None)
```

```python
data
```

```python
features=np.array(data.drop([64],axis=1))
target=np.array(data[64])
```

```python
features
```

```python
target=target.reshape(len(target),1)
```

```python
target
```

```python
mat=features[38].reshape(8,8)
print(target[38])
plt.imshow(mat,cmap='Greys')
```

```python
features=(features-features.min()/features.max()-features.min())
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(features,target,test_size=0.2)  
```

```python

model=keras.Sequential([
        keras.layers.Dense(81,activation=tf.nn.tanh,input_shape=(xtrain[1].shape)),
        keras.layers.Dense(10,activation=tf.nn.softmax)  
      
    ])
```

```python
adam=keras.optimizers.Adam(lr=0.001)
model.compile(loss='sparse_categorical_crossentropy',optimizer=adam,metrics=['accuracy'])
```

```python
history=model.fit(xtrain,ytrain,epochs=10,validation_split=0.2,verbose=2)
```

```python
plt.plot(history.epoch,history.history['acc'],'g')
plt.plot(history.epoch,history.history['val_acc'],'r')
```

```python
p=model.predict(xtrain[:3])
for i in p:
    print(np.argmax(i))
```

```python
model.evaluate(xtest,ytest)
```
